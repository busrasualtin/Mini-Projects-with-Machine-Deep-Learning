```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**Temel Cizim Ä°slemleri**


Plot y versus x as lines and/or markers. plot(*args[,scale_x,scale_y,data])

```python
plt.plot([1,2,3,4],[1,4,9,16])
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y)
plt.show()    # bu fonksiyon grafigin uzerinde cikan mesaji silecektir
```

```python
plt.title("First Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.show()
```

```python
# x ye y eksenlerindeki sayisal degerlerin dÃ¼zenini biz ayarlayabiliyoruz

plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title("First Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.show()
```

**Lejant Koyma**

```python
# buradaki fark label="x^2" ayÄ±racÄ±nÄ± ve plt.legend() kullanmak
# grafigin sag/sol ust kosesinde olusacak olan ifade icin kullanilir
plt.plot(x,y,label="x^2",color="green")
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title("First Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
# cizginin stiliyle oynamak icin

plt.plot(x,y,label="x^2",color="red", linewidth=2, linestyle='--', marker=".")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,1,4,9,16,25])
plt.title("First Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

**Ayni Grafikte Birden Fazla Cizim Yapmak**

```python
plt.plot(x,y,label="x^2",color="red", linewidth=2, linestyle='--', marker=".")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,1,4,9,16,25])
plt.title("First Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# ikinci cizim
x2=np.arange(0,5,0.5)
plt.plot(x2,x2*2,color="blue",linewidth=2, marker=".",label="2*x")

plt.legend()
plt.show()
```

**Grafigi Bilgisayara Kaydetme**

```python
plt.plot(x,y,label="x^2",color="red", linewidth=2, linestyle='--', marker=".")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,1,4,9,16,25])
plt.title("First Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# ikinci cizim
x2=np.arange(0,5,0.5)
plt.plot(x2,x2*2,color="blue",linewidth=2, marker=".",label="2*x")

plt.legend()
plt.savefig("firstgraph.png",dpi=300)
plt.show()
```

**Barchart**

```python
x=["Ankara","Istanbul","Izmir"]
y=[120,178,87]

plt.bar(x,y)
plt.show()
```

```python
# eger barlardan birini farklÄ± stille gorsellemek istiyorsak
bars=plt.bar(x,y)
bars[1].set_hatch("/")
bars[0].set_hatch(".")
plt.show()
```

**Ornekler**

```python
gas=pd.read_csv("petrolfiyatlari.csv")
gas
```

```python
plt.title("Petrol FiyatlarÄ±")
plt.plot(gas["Year"],gas["USA"],"b-",label="USA")

plt.xlabel("YÄ±l")
plt.ylabel("Dolar")

plt.legend()
plt.show()

# "b-" anlami cizgi seklinde blue olarak goster demek
```

```python
plt.title("Petrol FiyatlarÄ±")
plt.plot(gas.Year, gas.USA, "b-", label="USA")
plt.plot(gas.Year, gas.Canada, "r-", label="Canada")
plt.plot(gas.Year, gas["South Korea"], "g.-", label="South Korea")
plt.plot(gas.Year, gas.France, "y.-", label="France")

plt.xlabel("YÄ±l")
plt.ylabel("Dolar")

plt.legend()
plt.show()

# South Korea icin arada bosluk oldugundan dolayÄ± kÃ¶seli parantez yazim seklini uyguluyoruz.
# Diger ulkeler icin tek kelime isimlendirmeden dolayÄ± nokta yazÄ±m sekliyle kullanabiliyoruz.
```

**Figure Size**

```python
plt.figure(figsize=(18,7))
plt.title("Petrol FiyatlarÄ±")
plt.plot(gas.Year, gas.USA, "b-", label="USA")
plt.plot(gas.Year, gas.Canada, "r-", label="Canada")
plt.plot(gas.Year, gas["South Korea"], "g.-", label="South Korea")
plt.plot(gas.Year, gas.France, "y.-", label="France")

plt.xlabel("YÄ±l")
plt.ylabel("Dolar")

plt.legend()
plt.show()
```
