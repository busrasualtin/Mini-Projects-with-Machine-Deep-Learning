**Decision Tree Classification**

Bu Machine Learning modelinde karar verme yapısı oluşturulurken flowchart'lar yani akış diagramları kullanılır. Bu akış diagramına göre (leaf) yaprak'ler oluşturulur

Supervised Learning konsepti kapsamındadır. Sonuçta bir tree oluşturur.

```python
import pandas as pd
import numpy as np
from sklearn import tree

df=pd.read_csv("DecisionTreesClassificationDataSet.csv")
df
```

scikit-learn kütüphanesi decision tree'lerin düzgün çalışması için her şeyin rakamsal olmasını bekler. Bu nedenle veri setimizdeki tüm Y ve N değerlerini 0 ve 1 olarak düzeltiyoruz. Aynı sebeple eğitim seviyesini de BS:0, MS:1 ve PhD:2 olarak güncelliyoruz. map() kullanarak boş hücreler veya geçersiz değer girilen hücreler NaN ile doldurulacaktır.

```python
mapping={"Y":1,"N":0}

df["SuanCalisiyor?"]=df["SuanCalisiyor?"].map(mapping)
df["Top10 Universite?"]=df["Top10 Universite?"].map(mapping)
df["StajBizdeYaptimi?"]=df["StajBizdeYaptimi?"].map(mapping)
df["IseAlindi"]=df["IseAlindi"].map(mapping)

mapping2={"BS":0, "MS":1, "PhD":2}
df["Egitim Seviyesi"]=df["Egitim Seviyesi"].map(mapping2)

df
```

Sonuç sütununu ayırıyoruz.

```python
y=df['IseAlindi']
x=df.drop(["IseAlindi"],axis=1)
```

```python
x
```

DECISION TREE OLUŞTURALIM ...

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x,y)
```

```python
# Prediction yapalım..
# 5 yıl deneyimli, halihazırda bir yerde çalışan ve 3 eski şirkette çalışmış olan, eğitim seviyesi lisans
# top 10 üniversite mezunu değil, staj bizde yapmamış.

print(clf.predict([[5,1,3,0,0,0]]))
```

```python
# toplam 2 yıllık iş deneyimi, 7 kez iş değiştirmiş, iyi okul mezunu, şuan çalışmıyor.
print(clf.predict([[2,0,7,0,1,0]]))
```
