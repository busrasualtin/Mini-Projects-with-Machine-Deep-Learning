```python
df_filtered=df[df['IMDB_Rating']>=9]
df_filtered
```

```python
# veya...
df_filtered=df.query("IMDB_Rating>=9.2")
df_filtered
```

```python
# 140 dakikadan uzun filmleri gostermek istiyorum
# Sutunda '140 min' seklinde yaziyor yani int degil
# bu yuzden birtakim duzeltmeler yapmamiz gerekmektedir

# split() icine bir sey yazilmaz ise bosluga gore split ediliyor, yani 142 min yazısını 142 ve min olarak ikiye ayiriyor
# [0] eklememizin sebebi biz bu split edilmiş yerin sifirinci indexini yani sol taraftaki sayisal degerini almak istiyoruz
# bunu df.apply ile tum dataframe'e uyguluyorum
# axis=1 derken sutun degistirme islemi yapiyorum bu da RuntimeYeni adli yeni sutunum

f=lambda x: (x["Runtime"].split())[0]
df["RuntimeYeni"]=df.apply(f,axis=1)
```

```python
df.head()
#en sonda eklenmis yeni sutuna bakabilirsiniz...
```

```python
df.dtypes
```

```python
df["RuntimeYeni2"]=df["RuntimeYeni"].astype('int')
df_filtered=df.query("RuntimeYeni2>=140")
df_filtered
```
