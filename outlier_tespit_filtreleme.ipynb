**Outlier (Aykiri Deger) Tespiti ve Veri Setinden Temizlenmesi**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("outlier_ornek_veriseti.csv",sep=";")
df
```

```python
df.describe()
```

```python
# Q1 (%25 percentile hesaplama)
Q1=df.boy.quantile(0.25)
Q1
```

```python
# Q3 (%75 percentile hesaplama)
Q3=df.boy.quantile(0.75)
Q3
```

```python
# Median
Q2=df.boy.quantile(0.5)
Q2
```

```python
IQR=Q3-Q1
IQR
```

```python
# Alt Limit ve Ust Limit Hesaplama
alt_limit=Q1-(1.5)*IQR
ust_limit=Q3+(1.5)*IQR
```

```python
alt_limit
```

```python
ust_limit
```

```python
# Simdi bu hesaplamalari yaptigimiza gore verisetindeki aykiri degerleri bulabiliriz

df[(df.boy<alt_limit)|(df.boy>ust_limit)]
```

Bu degerler 'outlier' degerleridir. Istatiksel hesaplamalarimizin (ve tabiki bir cok yapay zeka algoritmasinin) daha verimli ve etkin calisabilmesi ve dogru sonuclar uretebilmesi icin outlier degerleri verisetimizden cikartacagiz.

**Outlier Degerlerin Verisetinden Cikartilmasi**
